```python
str = input("Numbers=? ")
strlist = str.split()

max = int(strlist[0])

for str1 in strlist[1:] :
  tmp = int(str1)
  if(tmp > max ):
    max = tmp

print("Max = ", max)
```

```python
def encode(sen, key):
  list = sen.split()
  list.reverse()
  reversedStr = " ".join(list)
  return conv_asc(reversedStr, key)

def decode(reversedStr, key):
  str = conv_asc(reversedStr, key*(-1))
  list = str.split()
  list.reverse()
  str = " ".join(list)
  return str


def conv_asc(reversedStr, key):
  converted_str = ""
  for ch in reversedStr:
    code = ord(ch)
    converted_str += chr(key+code)
  return converted_str
  

#main
sen = input("문자열을 입력하세요: ")
encoded = encode(sen, 1)
print("\""+sen+"\"is encoded into \""+encoded+"\"")
decoded = decode(encoded, 1)
print("\""+encoded+"\"is decoded into \""+decoded+"\"")
```
