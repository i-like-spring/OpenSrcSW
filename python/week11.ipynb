```python
f = open("README.txt", "r", encoding="UTF-8") #파일 경로에 주의

dic = dict()
for line in f:
  line = line.strip()
  words = line.split() # 리스트로 들어감~

  for w in words:
    if w in dic: # in 뒤에 리스트, 튜플, 딕셔너리가 존재~
      dic[w] +=1
    else:
      dic[w] = 1

for w in dic:
  print(w, dic[w]) #출력방식~

f.close()
```

```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y
  def move(self, x, y):
    self.x += x
    self.y +=y
  def show(self):
    #print((self.x, self.y))
    print("({0:d},{1:d})".format(self.x, self.y))
  
class Box(Point):
  def __init__(self, p1, p2):
    self.p1 = p1
    self.p2 = p2
  def move(self, dx, dy):
    self.p1.move(dx,dy)
    self.p2.move(dx,dy)
  def show(self):
    print("({0:d},{1:d}),({2:d},{3:d})".format(self.p1.x, self.p1.y, self.p2.x, self.p2.y))  #여기도 출력 포맷을 이쁘게~

print("Point")
p = Point(1,1)
p.show()
p.move(1,2)
p.show()
print("Box")
b = Box(Point(1,1), Point(2,2))
b.show()

  
```
